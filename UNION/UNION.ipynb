## Norm

```python
indexes = [x for x in list(movies.index) if x not in range(101, 251) and x not in [42, 48, 252, 62, 99, 93]]
movies = movies.loc[indexes]
```

```python
def norm(series):
    return (series - np.average(series[~series.isnull()])) / np.std(series[~series.isnull()])

def norm2(series):
    return (series - 40) / np.std(series)
```

```python
movies['norm_brightness'] = norm(movies['brightness'])
movies['norm_emotions h/(h+s)'] = norm(movies['emotions h/(h+s)'])
movies['norm_emotions h/all'] = norm(movies['emotions h/all'])
movies['norm_emotions s/all'] = norm(movies['emotions s/all'])
movies['norm_nsfw'] = norm(movies['nsfw'])
movies['norm_gunshots'] = norm2(movies['gunshots'])

movies = movies.dropna()
```

```python
movies['id'] = movies.index
```

```python
def movie_score(movie_params, user_happiness, user_bullets, user_brightness, user_sexuality):
    return 1./2 * 1./movie_params['id'] + \
           user_happiness * movie_params['norm_emotions h/(h+s)'] + \
           user_brightness * movie_params['norm_brightness'] + \
           user_sexuality * movie_params['norm_nsfw'] + \
           1.2 * user_bullets * movie_params['norm_gunshots']
```

```python
def get_top10(user_happiness, user_bullets, user_brightness, user_sexuality):
    scores = [movie_score(row, user_happiness, user_bullets, user_brightness, user_sexuality) for idx, row in movies.iterrows()]
    top10_idx = np.argsort(scores)[::-1][:10]
    
    for idx in top10_idx:
        print(scores[idx])
    
    return top10_idx
```

```python
user_happiness, user_bullets, user_brightness, user_sexuality = 1, 0, 1, 1
top10 = get_top10(user_happiness, user_bullets, user_brightness, user_sexuality)
movies.iloc[top10]
```
